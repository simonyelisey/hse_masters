## Задание 2

Найдите экстремумы этих функций в Python.

### Решение

```python
import sympy as sp
```

```python
x = sp.Symbol('x')

# определим функции

def f_f(x):

  result = x**2 - 30*x +1

  return result

def f_g(x):

  result = -x**3 + 5*x -7*x**2 +17

  return result

def f_h(x):

  result = x**2 - 1/x

  return result

def f_j(x):

  result = 1/x**5 + x**17 +8

  return result
```

```python
# первые производные функций
df_f = sp.diff(f_f(x))
df_g = sp.diff(f_g(x))
df_h = sp.diff(f_h(x))
df_j = sp.diff(f_j(x))

# точки экстремума
ext_f = sp.solveset(df_f, domain=sp.Reals)
ext_g = sp.solveset(df_g, domain=sp.Reals)
ext_h = sp.solveset(df_h, domain=sp.Reals)
ext_j = sp.solveset(df_j, domain=sp.Reals)
```

```python
for func, ext in zip(
    ['f', 'g', 'h', 'g'],
    [ext_f.args, ext_g.args, ext_h.args, ext_j.args]
    ):
  print(f'Точка/и экстремума функции {func}: {ext}\n')
```

```python
# определяем тип экстремум
for func, ext in zip([
    ('f', df_f), ('g', df_g), ('h', df_h), ('j', df_j)],
    [ext_f.args, ext_g.args, ext_h.args, ext_j.args
     ]):
  # считаем вторую производную
  second_diff = sp.diff(func[1])
  # проверяем второе достаточное условие
  for point in ext:
    diff_at_point = second_diff.subs(x, point)
    if diff_at_point < 0:
      print(f'{func[0]}(x) = {point} - точка максимума функции.')
    elif diff_at_point > 0:
      print(f'{func[0]}(x) = {point} - точка минимума функции.')
    else:
      print(f'{func[0]}(x) = {point} - данная точка может не быть точкой \
       экстремума, исследуйте функцию дальше.')
  print('\n')
```

```python
def plot_the_function(
    func, xlim, ylim, extremes=False, inflection=False, gradient_descent=False
    ):
  """
  Функция отрисовыват графики функции
  """
  plt.figure(figsize=(8, 8))

  # значения функции
  x_values = np.linspace(-50, 50, 10000)
  y_values = [func[0](i) for i in x_values]

  # график цункции
  plt.plot(x_values, y_values)

  if gradient_descent:
      label='точка минимума, найденная градиентныйм спуском'
  else:
      label='точки экстремума'
  # точки экстремума и перегиба, если они есть
  if extremes:
    plt.scatter(
      extremes,
      [func[0](i) for i in extremes],
      c='red',
      label=label
      )
  if inflection:
    plt.scatter(
        inflection,
        [func[0](i) for i in inflection],
        c='green',
        label='точки перегииба'
        )
  # оси
  plt.axvline(x=0, c = 'black', alpha=0.5)
  plt.axhline(y=0, c = 'black', alpha=0.5)  

  # подпись графика фукнкции
  plt.xlabel('x')
  plt.ylabel(f'{func[1]}')
  if gradient_descent:
    plt.title(f'График функции {func[1]} с точкой минимума, найденной градиентным спуском.')
  else:
    plt.title(f'График функции {func[1]} с точками экстремум и перегиба.')

  # анстройки отображения
  plt.xlim(xlim)
  plt.ylim(ylim)
  plt.legend()
```

```python
def annotate_points(points, func, xshift=0, yshift=0):
  """
  Функция производит аннотацию точек на графике функции
  """
  for i in points:
    plt.annotate(
        (round(i, 2), round(func(i), 2)),
        xy=(i, func(i)),
        xytext=(i+xshift, func(i)+yshift)
        )
```

## Задание 6

Модифицируйте функцию, осуществляющую метод градиентного спуска с семинара: добавьте опцию останавливать метод, если абсолютное значение производной функции в точке < некоторого $\delta$ (delta добавьте в аргументы функции).

```python
def gradient_descent(x_start, learning_rate, epsilon, num_iterations, f, delta):
    # начальная точка
    x_curr = x_start
    x = sp.Symbol('x')
    df_x = sp.diff(f(x)) # производная

    trace = []
    trace.append(x_curr)

    for i in range(num_iterations):
      # формула градиентного спуска
      x_new = x_curr - learning_rate * df_x.subs(x, x_curr) 
      trace.append(x_new)

      if np.abs(df_x.subs(x, x_curr)) < delta:
        return x_curr, trace

      if np.abs(x_new - x_curr) < epsilon:
        return x_curr, trace

      x_curr = x_new

    return x_curr, trace
```

## Задание 8

Реализуйте [метод Ньютона](https://translated.turbopages.org/proxy_u/en-ru.ru.c97ef85d-631c4b6a-ff938028-74722d776562/https/en.wikipedia.org/wiki/Newton%27s_method_in_optimization) (метод второго порядка) для поиска минимума функции.

Про метод ещё можно почитать [здесь](https://math.semestr.ru/optim/minimum.php).

Формула метода Ньютона:

$x_{new} = x - \frac{f'(x)}{f''(x)}$

```python
def gradient_descent_nt(x_start, epsilon, num_iterations, f, delta):
    # начальная точка
    x_curr = x_start
    x = sp.Symbol('x')
    df_x = sp.diff(f(x)) # производная
    ddf_x = sp.diff(df_x) # 2 производная

    trace = []
    trace.append(x_curr)

    for i in range(num_iterations):
      # градиентный спуск по методу Ньютона
      x_new = x_curr - df_x.subs(x, x_curr) / ddf_x.subs(x, x_curr)
      trace.append(x_new)

      if np.abs(df_x.subs(x, x_curr)) < delta:
        return x_curr, trace

      if np.abs(x_new - x_curr) < epsilon:
        return x_curr, trace

      x_curr = x_new

    return x_curr, trace
```

```python
# списки для хранения количество итераций, необходимых для схождения к минимуму

classic_all_function_iters = []
newton_all_function_iters = []

classic_differences = []
newton_differences = []
```

```python
def compare_gd_methods(func, x_start, lr, local_min):
  """
  Функция итеративно подбирает количестко итераций необходимых для схождения к минимуму
  """
  # классический метод
  iters_classic = 0
  xmin = -999
  while round(xmin, 6) != round(local_min, 6):
    xmin, trace = gradient_descent(
        x_start=x_start,
        learning_rate=lr,
        epsilon=0,
        num_iterations=iters_classic,
        f=func,
        delta=0
        )
    iters_classic += 1
  
  # метод Ньютона
  iters_newton = 0
  xmin = -999
  while round(xmin, 6) != round(local_min, 6):
    xmin, trace = gradient_descent_nt(
        x_start=x_start,
        epsilon=0,
        num_iterations=iters_newton,
        f=func,
        delta=0
        )
    iters_newton += 1

  return iters_classic, iters_newton


def compare_fixed_iter_gd_methods(func, x_start, lr, local_min):
  """
  Функция производит градиентный спуск с фиксированным количеством итераций
  """
  xmin_classic, trace_classic = gradient_descent(
        x_start=x_start,
        learning_rate=lr,
        epsilon=0,
        num_iterations=10,
        f=func,
        delta=0
        )
  
  xmin_newton, trace_newton = gradient_descent_nt(
        x_start=x_start,
        epsilon=0,
        num_iterations=10,
        f=func,
        delta=0
        )
  
  classic_diff = round(xmin_classic, 6) - round(local_min, 6)
  newton_diff = round(xmin_newton, 6) - round(local_min, 6)

  return classic_diff, newton_diff
```

## Задание 11*

Реализуйте оптимизацию метода градиентного спуска - [Momentum](https://ml-handbook.ru/chapters/optimization/intro#метод-инерции-momentum).

```python
def gradient_descent_momentum(
    x_start, learning_rate, epsilon, num_iterations, f, delta, betta):
    # начальная точка
    x_curr = x_start
    x = sp.Symbol('x')
    df_x = sp.diff(f(x)) # производная

    trace = []
    trace.append(x_curr)

    for i in range(num_iterations):
      if len(trace) < 2:
        # формула градиентного спуска
        x_new = x_curr - learning_rate * df_x.subs(x, x_curr)
      else:
        # реализация метода momentum
        x_new = x_curr - (learning_rate * df_x.subs(x, x_curr) +
                          betta * (x_curr - trace[-2]))
      trace.append(x_new)

      if np.abs(df_x.subs(x, x_curr)) < delta:
        return x_curr, trace

      if np.abs(x_new - x_curr) < epsilon:
        return x_curr, trace

      x_curr = x_new

    return x_curr, trace
```
